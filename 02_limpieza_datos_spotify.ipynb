### Limpieza de Datos

```python
import pandas as pd
```

```python
data = pd.read_csv('spotify_dataset.csv')

data
```

### PASO 1: Eliminación de columnas irrelevantes

Eliminamos 'Unnamed: 0' ya que es un índice duplicado que no aporta información.

```python
df = data.drop(columns=['Unnamed: 0'])

df
```

### PASO 2: Tratamiento de valores nulos

Como vimos que solo hay 1 fila con nulos en artistas, álbum y canción, 
la eliminamos directamente ya que representa una pérdida insignificante.

```python
df = df.dropna(subset=['artists', 'album_name', 'track_name'])

df
```

Comprobación de celdas vacías (strings vacíos, no nulos).


isnull() no detecta '' ni ' ' : buscamos esos casos aparte.

```python
# Obtener todas las columnas del DataFrame para la comprobación

all_columns = df.columns

print('Celdas con string vacío o solo espacios por columna (todas las columnas):')
for col in all_columns:
    # Se convierte a string para asegurar que str.strip() funcione en todos los tipos de datos
    n = (df[col].astype(str).str.strip() == '').sum()
    if n > 0:
        print(f'  ⚠ {col}: {n} celdas vacías')
        # Mostrar las primeras filas donde se encuentren estos casos
        display(df[df[col].astype(str).str.strip() == ''][[col]].head())
    else:
        print(f'  ✓ {col}: sin celdas vacías')
```

Limpiar espacios al inicio y al final en columnas de texto. 


str.strip() elimina espacios invisibles que pueden causar errores en agrupaciones

```python
for col in ['artists', 'album_name', 'track_name']:
    df.loc[:, col] = df[col].str.strip()
    print(f'  ✓ {col}: espacios limpiados')
```

### PASO 3: Tratamiento de valores duplicados

```python
pd.DataFrame(df['track_id']).groupby('track_id').agg('count')
```

Investigar duplicados por track_id: cuáles son y por qué. No basta con contarlos, miramos un caso concreto para entender la causa.

```python
repeticiones = df.groupby('track_id').size().reset_index(name='veces')
repetidos = repeticiones[repeticiones['veces'] > 1].sort_values('veces', ascending=False)

print(f'track_id que aparecen más de 1 vez: {len(repetidos)}')
print('\nDistribución de repeticiones:')
display(repeticiones['veces'].value_counts().rename_axis('apariciones').reset_index(name='nº de tracks'))

# Caso concreto con más repeticiones
if not repetidos.empty:
    ejemplo_id = repetidos.iloc[0]['track_id']
    n_veces = repetidos.iloc[0]['veces']
    print(f"\nEjemplo — track_id '{ejemplo_id}' aparece {n_veces} veces:")
    display(df[df['track_id'] == ejemplo_id][['track_id', 'track_name', 'artists', 'track_genre', 'popularity']])
    print('\n→ Conclusión: la misma canción aparece una vez por cada género en el que')
    print('  está clasificada en Spotify. Es la estructura del dataset, no un error.')
else:
    print('\nNo hay track_id repetidos para investigar después de la limpieza.')
```

Eliminamos duplicados exactos (filas idénticas si las hubiera)

```python
df = df.drop_duplicates()

# Después, se eliminan los registros que tengan el identificador de canción ('track_id') repetido,
# manteniendo únicamente la primera aparición detectada (keep='first')

# df = df.drop_duplicates(subset=['track_id'], keep='first') 

# NO HAY QUE BORRARLOS SI ES LA ESTRUCTURA DEL DATASET Y NO UN ERROR

df
```

### PASO 4: Optimización y corrección de tipos de datos

```python
# A) Convertir columnas de texto repetitivo a tipo 'category'
df.loc[:, 'track_genre'] = df['track_genre'].astype('category')
df.loc[:, 'key'] = df['key'].astype('category')
df.loc[:, 'mode'] = df['mode'].astype('category')
df.loc[:, 'time_signature'] = df['time_signature'].astype('category')

# B) Reducir enteros de 64 bits a tamaños más eficientes
df.loc[:, 'popularity'] = df['popularity'].astype('int8')     # Rango 0 a 100 cabe en int8
df.loc[:, 'duration_ms'] = df['duration_ms'].astype('int32')   # Duración cabe en int32

# C) Reducir flotantes de 64 bits a 32 bits (precisión suficiente)
columnas_flotantes = [
    'danceability', 'energy', 'loudness', 'speechiness', 
    'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
]
# Para modificar múltiples columnas a la vez con .loc, pasamos la lista de columnas
df.loc[:, columnas_flotantes] = df[columnas_flotantes].astype('float32')

# D) Convertir columnas 'object' (que son texto) a tipo 'string'
object_cols = df.select_dtypes(include='object').columns
df.loc[:, object_cols] = df[object_cols].astype('string')
```
